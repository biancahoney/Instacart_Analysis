# Data wrangling

```python
# Rename columns 

df_customers.rename(columns = {'First Name' : 'first_name'}, inplace = True)

df_customers.rename(columns = {'Surnam' : 'surname'}, inplace = True)

df_customers.rename(columns = {'Gender' : 'gender'}, inplace = True)

df_customers.rename(columns = {'STATE' : 'state'}, inplace = True)

df_customers.rename(columns = {'Age' : 'age'}, inplace = True)
```

```python
df_customers.head(5)
```

# Find Missing Values 

```python
df_customers.isnull().sum()
```

```python
# Create a subset of missing values
df_nan = df_customers[df_customers['First Name'].isnull == True]
```

```python
df_nan
```

```python
# No real pattern or alternative for missing values - replace with 'unknown'

df_customers['first_name'].fillna('Unknown', inplace=True)
```

```python
df_customers.isnull().sum()
```

```python
# Search for duplicates

df_dups = df_customers[df_customers.duplicated()]
```
