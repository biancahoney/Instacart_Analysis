### 2.  groupby and agg() function

```python
# group data by specific column name

df.groupby('product_name')
```

```python
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# the above differs from previously calculated subset means in the sense that all department Id's are now accounted for and the previiously calculated means (as part of the subset) have decreased slightly.
```

### 2a. Create loyalty flag
   Aggregate data with transform

```python
# Creating loyatly flag for customers

df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df.head(100)
```

### 2b. Deriving columns with loc function

```python
# Deriving columns with loc funtion

df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df['loyalty_flag'].value_counts(dropna = False)
```

```python
df[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
df.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# Calculate the average price per user using transform

df['average_price'] = df.groupby('user_id')['prices'].transform(np.mean)
```

### 2c. Create Spending flags 

```python
# Create spending flags based on average prices

df['spending_flag'] = 'Low spender'
```

```python
df.loc[df['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df.head(100)
```

```python
# Determining frequent vs. non-frequent customers

df['median_days'] = df.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
df.head(5)
```

### 2d. Create frequency flags 

```python
# Create order frequency flags

df.loc[df['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df.loc[(df['median_days'] > 10) & (df['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df.loc[df['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequencies of spending flags for accuracy

df['frequency_flag'].value_counts(dropna=False)
```

```python
# Check null values

print(df[df.frequency_flag.isnull()])
```

```python
# median cannot be created for customer as they have only had 1 order, therefore can qualify as non frequent customer
```

```python
df['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
df['frequency_flag'].value_counts(dropna=False)
```
